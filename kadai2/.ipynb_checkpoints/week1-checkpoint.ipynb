```python
result = reserve_tb.groupby('hotel_id')['customer_id'].nunique().reset_index()

result.columns = ['hotel_id','cus_cnt'] #列名を変更
result.head()
```

```python
result = reserve_tb.groupby('hotel_id').agg({'reserve_id':'count','customer_id':'nunique'})
result.reset_index(inplace=True)

result.columns = ['hotel_id','rsv_cnt','cus_cnt']#列名を変更
result.head()
```

```python
result = reserve_tb.groupby(['hotel_id','people_num'])['total_price'].sum().reset_index()

result.rename(columns={'total_price':'price_sum'},inplace=True)
result.head()
```

```python
result = reserve_tb.groupby('hotel_id').agg({'total_price':['max','min','mean','median']}).reset_index()

result.columns = ['hotel_id', 'price_max','price_min','price_mean','price_median']
result.head()
```

```python
result = reserve_tb.groupby('hotel_id').agg({'total_price':['var','std']}).reset_index()
result.columns = ['hotel_id','price_var','price_std']

result.fillna(0,inplace=True)
result.head()
```

```python
rsv_cnt_tb = reserve_tb.groupby('hotel_id').size().reset_index()

rsv_cnt_tb.columns = ['hotel_id','rsv_cnt']

rsv_cnt_tb['rsv_cnt_rank'] = rsv_cnt_tb['rsv_cnt'].rank(ascending=False, method='min')
rsv_cnt_tb.drop('rsv_cnt',axis=1,inplace=True) #必要のないrsv_cntの列を削除
rsv_cnt_tb.head()
```

```python
rsv_cnt_tb.sort_values('rsv_cnt_rank',ascending=True)
```

```python
reserve_tb['price_mean'] = pd.Series(reserve_tb.groupby('customer_id').apply(lambda x: x.sort_values(by='reserve_datetime',ascending=True))
                                    .loc[:,'total_price'].rolling(center=False,window=3,min_periods=1).mean().reset_index(drop=True))

reserve_tb.head()
```

```python
#課題2
import numpy as np
import pandas as pd
reserve_tb = pd.read_csv('data/reserve.csv', encoding='UTF-8')
customer_tb = pd.read_csv('data/customer.csv',encoding='UTF-8')

kadai2_rslt = pd.merge(reserve_tb.query('"2016-01-01" <= reserve_datetime <= "2016-12-31"'),customer_tb.query('sex == "man"'),
                       on='customer_id',how='inner')
kadai2_rslt = kadai2_rslt.groupby('hotel_id').agg({'total_price':['max']}).reset_index()
kadai2_rslt.columns = ['hotel_id','price_max']
kadai2_rslt['price_max_rank'] = kadai2_rslt['price_max'].rank(ascending=False,method='min')
kadai2_rslt = kadai2_rslt.sort_values('price_max_rank', ascending=True)
kadai2_rslt.head(10)
```

```python
##課題1(reserve_tbの行を50％サンプリングした後に顧客を重複なく列にした場合)
import numpy as np
import pandas as pd
reserve_tb = pd.read_csv('data/reserve.csv', encoding='UTF-8')

reserve_sample = reserve_tb.sample(frac=0.5)
target = reserve_sample['customer_id'].unique()

kadai1_rslt = reserve_tb[reserve_tb['customer_id'].isin(target)].groupby('customer_id').size().reset_index()

kadai1_rslt.columns = ['customer_id','rsv_cnt']

print(kadai1_rslt['rsv_cnt'])
```

```python
##課題1(reserve_tbから重複のない顧客の列を抽出し50％サンプリングをした場合)
import numpy as np
import pandas as pd
reserve_tb = pd.read_csv('data/reserve.csv', encoding='UTF-8')

target = pd.Series(reserve_tb['customer_id'].unique()).sample(frac=0.5)
kadai1_rslt = reserve_tb[reserve_tb['customer_id'].isin(target)].groupby('customer_id').size().reset_index()

kadai1_rslt.columns = ['customer_id','rsv_cnt']

print(kadai1_rslt['rsv_cnt'])
```

```python
#課題3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
mmsc = MinMaxScaler()
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
from sklearn.datasets import load_iris
iris = load_iris()

x1 = iris.data[:,0].reshape(-1,1)
x2 = iris.data[:,1].reshape(-1,1)

x=np.concatenate((x1,x2), axis=1)
plt.axis('equal')
plt.scatter(x[:,0],x[:,1], marker="x", label="org")
x_norm = mmsc.fit_transform(x)
x_std = stdsc.fit_transform(x)
plt.scatter(x_std[:,0],x_std[:,1], c='red',marker="x", label="std")
plt.scatter(x_norm[:,0],x_norm[:,1], c='green',marker="x", label="norm")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.grid()
plt.show()
```

```python
##課題4
import numpy as np
import pandas as pd
reserve_tb = pd.read_csv('data/reserve.csv', encoding='UTF-8')

reserve_tb['reserve_datetime'] = pd.to_datetime(reserve_tb['reserve_datetime'])

kadai4_rslt = pd.merge(reserve_tb.loc[:, ['reserve_id','customer_id','reserve_datetime']],\
                    reserve_tb.loc[:, ['customer_id','reserve_datetime', 'total_price']],on='customer_id',how='inner')
kadai4_rslt['diffday'] = (kadai4_rslt['reserve_datetime_x'] - kadai4_rslt['reserve_datetime_y']).dt.days
kadai4_rslt = kadai4_rslt.query('0 < diffday <= 90').groupby('reserve_id')['total_price'].mean().reset_index()
kadai4_rslt.columns = ['reserve_id', 'total_price_90d']
kadai4_rslt = pd.merge(reserve_tb,kadai4_rslt,on='reserve_id',how='left').fillna(0)
kadai4_rslt.head(15)
```
